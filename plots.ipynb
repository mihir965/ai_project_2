```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r'Comparison_Summary.xlsx' 
data = pd.read_excel(file_path, sheet_name='Sheet1')
```

```python
# Stationary rat case (Improved Bot vs Baseline Bot Accuracy as a function of alpha)

plt.figure(figsize=(10, 6))
plt.plot(data['Alpha'], data['Baseline Accuracy'], label='Baseline Accuracy', marker='o')
plt.plot(data['Alpha'], data['Improved Accuracy'], label='Improved Accuracy', marker='x')
plt.title('Performance Comparison (Stationary Rat)')
plt.xlabel('Alpha')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Moving rat case (Improved Bot vs Baseline Bot Move Accuracy as a function of alpha)

plt.figure(figsize=(10, 6))
plt.plot(data['Alpha'], data['Baseline Move Accuracy'], label='Baseline Move Accuracy', marker='o')
plt.plot(data['Alpha'], data['Improved Move Accuracy'], label='Improved Move Accuracy', marker='x')
plt.title('Performance Comparison (Moving Rat)')
plt.xlabel('Alpha')
plt.ylabel('Move Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
